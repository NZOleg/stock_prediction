```python
from stockstats import StockDataFrame as Sdf
import quandl
import pandas as pd
import numpy as np
```

```python
stock_list = pd.read_csv('WIKI-datasets-codes.csv')
stock_list = stock_list['WIKI/AAPL']
```

```python
for stock_index in stock_list:
    data =quandl.get(stock_index, authtoken="gpEsN15cRWP7-srvwW9X")
    data['LIBOR'] = quandl.get("FRED/USDONTD156N", authtoken="gpEsN15cRWP7-srvwW9X")
    data['Dollar_Index'] = quandl.get("FRED/DTWEXM", authtoken="gpEsN15cRWP7-srvwW9X")
    stock = Sdf.retype(data)
    # volume delta against previous day
    data['volume_delta'] = stock['volume_delta']

    # open delta against next 2 day
    data['open_2_d'] = stock['open_2_d']

    # open price change (in percent) between today and the day before yesterday
    # 'r' stands for rate.
    data['open_-2_r'] = stock['open_-2_r']

    # CR indicator, including 5, 10, 20 days moving average
    data['cr'] = stock['cr']
    data['cr-ma1'] = stock['cr-ma1']
    data['cr-ma2'] = stock['cr-ma2']
    data['cr-ma3'] = stock['cr-ma3']

    # volume max of three days ago, yesterday and two days later
    data['volume_-3,2,-1_max'] = stock['volume_-3,2,-1_max']

    # volume min between 3 days ago and tomorrow
    data['volume_-3~1_min'] = stock['volume_-3~1_min']

    # KDJ, default to 9 days
    data['kdjk'] = stock['kdjk']
    data['kdjd'] = stock['kdjd']
    data['kdjj'] = stock['kdjj']

    # three days KDJK cross up 3 days KDJD
    #data['kdj_3_xu_kdjd_3'] = stock['kdj_3_xu_kdjd_3']

    # 2 days simple moving average on open price
    data['open_2_sma'] = stock['open_2_sma']

    # MACD
    data['macd'] = stock['macd']
    # MACD signal line
    data['macds'] = stock['macds']
    # MACD histogram
    data['macdh'] = stock['macdh']

    # bolling, including upper band and lower band
    data['boll'] = stock['boll']
    data['boll_ub'] = stock['boll_ub']
    data['boll_lb'] = stock['boll_lb']

    # close price less than 10.0 in 5 days count
    data['close_10.0_le_5_c'] = stock['close_10.0_le_5_c']

    # CR MA2 cross up CR MA1 in 20 days count
    data['cr-ma2_xu_cr-ma1_20_c'] = stock['cr-ma2_xu_cr-ma1_20_c']

    # 6 days RSI
    data['rsi_6'] = stock['rsi_6']
    # 12 days RSI
    data['rsi_12'] = stock['rsi_12']

    # 10 days WR
    data['wr_10'] = stock['wr_10']
    # 6 days WR
    data['wr_6'] = stock['wr_6']

    # CCI, default to 14 days
    data['cci'] = stock['cci']
    # 20 days CCI
    data['cci_20'] = stock['cci_20']

    # TR (true range)
    data['tr'] = stock['tr']
    # ATR (Average True Range)
    data['atr'] = stock['atr']

    # DMA, difference of 10 and 50 moving average
    data['dma'] = stock['dma']

    # DMI
    # +DI, default to 14 days
    data['pdi'] = stock['pdi']
    # -DI, default to 14 days
    data['mdi'] = stock['mdi']
    # DX, default to 14 days of +DI and -DI
    data['dx'] = stock['dx']
    # ADX, 6 days SMA of DX, same as stock['dx_6_ema']
    data['adx'] = stock['adx']
    # ADXR, 6 days SMA of ADX, same as stock['adx_6_ema']
    data['adxr'] = stock['adxr']

    # TRIX, default to 12 days
    data['trix'] = stock['trix']
    # MATRIX is the simple moving average of TRIX
    data['trix_9_sma'] = stock['trix_9_sma']

    # VR, default to 26 days
    data['vr'] = stock['vr']
    # MAVR is the simple moving average of VR
    data['vr_6_sma'] = stock['vr_6_sma']
    def direction_up(row):
        if(row['change'] > 0):
            return 1
        else:
            return 0
    def direction_down(row):
        if(row['change'] < 0):
            return 1
        else:
            return 0
    def direction(row):
        return row['change']/100
    data['direction_up'] = data.apply(direction_up, axis=1)
    data['direction_down'] = data.apply(direction_down, axis=1)
    data['direction'] = data.apply(direction, axis=1)
    data['direction_up'] = data['direction_up'].shift(-1)
    data['direction_down'] = data['direction_down'].shift(-1)
    data['direction'] = data['direction'].shift(-1)
    data = data[np.isfinite(data['libor'])]
    data = data[np.isfinite(data['cr-ma1_20_c'])]
    data.to_csv('stock_full/'+stock_index.split('/')[1]+'.csv')
    print(stock_index.split('/')[1]+'.csv is saved');
```
